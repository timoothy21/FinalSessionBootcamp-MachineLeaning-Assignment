OOP

```python
class Animal:
  def walk(self):
    print("walking...")

class Cow(Animal):
  def __init__(self, name: str):
    self.__name = name

  def talk(self) -> None:
    print("{} Mooooooooooooooooooo".format(self.__name))

  def walk(self) -> None:
    print("This is a cow walking!")

class Cat(Animal):
  pass
```

```python
budi = Cow('sapi budi')
budi.walk()
budi.talk()
```

```python
meong = Cat()
meong.walk()
```
